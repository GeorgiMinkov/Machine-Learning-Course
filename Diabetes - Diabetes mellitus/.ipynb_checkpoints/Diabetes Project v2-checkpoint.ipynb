## Ще направим сега основен модел (изследване на суровите данни спрямо няколко модела)

```python
from sklearn.metrics import confusion_matrix
def conf_matrix_visualizer(model, X_test, y_test):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)

    conf_matrix = pd.DataFrame(data = cm, columns = ['Predicted:false','Predicted:true'], index = ['Actual:false','Actual:true'])
    plt.figure(figsize = (8,5))
    sns.heatmap(conf_matrix, annot = True, fmt = 'd') # ,cmap="YlGnBu"
```

<center> <h1> svm - support vector machine </h1> </center>

```python
kernels = ['linear', 'rbf']
for kernel in kernels:
    model = svm.SVC(kernel = kernel)
    model.fit(X_train, y_train)
    conf_matrix_visualizer(model, X_test, y_test)
    prediction = model.predict(X_test)
    print('Using kernel -> {} get prediction rate -> {}'.format(kernel, metrics.accuracy_score(prediction, y_test)))
```

Линейния модел има сравнително равномерно разпределение, когато трябва да предсказва коректно данните, но rbf има силен underfit

<center> <h1> Logistic Regression  </h1> </center>

```python
predictions = []
model = LogisticRegression()
model.fit(X_train, y_train)
predictions.append(metrics.accuracy_score(model.predict(X_train), y_train))
predictions.append(metrics.accuracy_score(model.predict(X_test), y_test))
print('Train score for logistic regression -> {}\nTest score for logistic regression -> {}'.format(predictions[0], predictions[1]))
```

```python
conf_matrix_visualizer(model, X_test, y_test)
```

<center> <h1> Decision tree  </h1> </center>

```python
model = DecisionTreeClassifier(criterion='gini')
model.fit(X_train, y_train)
predictions.append(metrics.accuracy_score(model.predict(X_train), y_train))
predictions.append(metrics.accuracy_score(model.predict(X_test), y_test))
print('Train score for logistic regression -> {}\nTest score for logistic regression -> {}'.format(predictions[0], predictions[1]))
```

Забелязва се overfit, въпреки ниския процен на train score-a

```python
conf_matrix_visualizer(model, X_test, y_test)
```

Двата модела стигат до сравнителни подобни резултати, когато трябва да се предскаже правилно дали един човек не е болен, то се справя с добра точност, но когато трябва да посочи отрицателния сценарии - да кажем, че няма , но всъчщност индивидът да има диабет, тогава моделът носи голяма грешка

<center> <h1> K-Nearest Neighbours </h1> </center>

Тук ще направим експеримент с променлив брой на K за да следим регуляризацията

```python
training_accuracy = []
test_accuracy = []
# try n_neighbors from 1 to 100
neighbors_settings = range(1, 101)

for n_neighbors in neighbors_settings:
    # build the model
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(X_train, y_train)
    # record training set accuracy
    training_accuracy.append(knn.score(X_train, y_train))
    # record test set accuracy
    test_accuracy.append(knn.score(X_test, y_test))

plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Number of neighbors")
plt.legend()
plt.savefig('knn_compare_model')
```

```python
# Сравнително стабилни стойностти имаме около 9 и 21
knn = KNeighborsClassifier(n_neighbors = 9)
knn.fit(X_train, y_train)
print ("Train score with knn = 9 -> {}\nTest score with knn = 9 -> {}".format(knn.score(X_train, y_train), 
                                                                               knn.score(X_test, y_test)))
```

```python
# with 21
knn = KNeighborsClassifier(n_neighbors = 21)
knn.fit(X_train, y_train)
print ("Train score with knn = 21 -> {}\nTest score with knn = 21 -> {}".format(knn.score(X_train, y_train), 
                                                                               knn.score(X_test, y_test)))
```

```python
# но имаме интересен резултат около 80
knn = KNeighborsClassifier(n_neighbors = 82)
knn.fit(X_train, y_train)
print ("Train score with knn = 82 -> {}\nTest score with knn = 82 -> {}".format(knn.score(X_train, y_train), 
                                                                               knn.score(X_test, y_test)))
```

Обобщено при ниски стойностти на параметъра набюдаваме силен overfit.
Три точки правят главно впечатление - 9 ; 21 ; 82

```python
conf_matrix_visualizer(knn, X_test, y_test)
```

Горепосочената диаграма представлява confusion matrix при натренирания с kNN модел. Моделът добре се справя когато трябва да предскаже, че даден субект няма диабет, но когато трябва да предскаже позитивен резултат изпитва проблеми.
Съотношението което ни дава false negative и false positive не е в полза на субекта, т.е ще предскажем, че даден човек не е болен, а всъщност е

Можем да обобщим горните набюдения в конкретна таблица, която да показва избраните класификатори

```python
def score_fit(features, labels):
    return metrics.accuracy_score(model.predict(features), labels)

def is_potential_overfit(results):
    return [(el[0] - el[1] > 0.07) for el in results]

def table(results, fitting):
    return [(el[0], el[1], fit) for (el, fit) in (results, fitting)]
    
```

```python
results=[]
classifiers=['Linear Svm','Radial Svm','Logistic Regression','KNN','Decision Tree']
models=[svm.SVC(kernel='linear'),svm.SVC(kernel='rbf'),LogisticRegression(),KNeighborsClassifier(n_neighbors=9),DecisionTreeClassifier()]
for i in models:
    model = i
    model.fit(X_train, y_train)
    results.append((score_fit(X_train, y_train), score_fit(X_test, y_test)))
    
over = is_potential_overfit(results)
results = list(zip(results, is_potential_overfit(results)))
results = [(el[0][0], el[0][1], el[1]) for el in results]
```

```python
# Сега ще ги визуализираме
models_dataframe=pd.DataFrame(results,index=classifiers)   
models_dataframe.columns=['Train Accuracy', 'Test Accuracy', 'Potential Overfit']
models_dataframe
```

## Standardisation

В данните може да се наблюдава отклонение в типа данни (въпреки, че са количествени). Пример -> BMI има над 200 уникални стойностти. Подобни отклонения би било добре да ги стандартизираме.Това би било подходящо за Гаусово разпределение и Поасоново разпределение.

## Стандартизиране по основните feature-и

```python
from sklearn.preprocessing import StandardScaler

new_data = matrix[['Glucose','BMI','Age','DiabetesPedigreeFunction','Outcome']]

features = new_data[new_data.columns[:4]]
features_standard = StandardScaler().fit_transform(features) # Gaussian Standardisation

x = pd.DataFrame(features_standard,columns=[['Glucose','BMI','Age','DiabetesPedigreeFunction']])
x['Outcome'] = new_data['Outcome']

outcome= x['Outcome']

train1, test1 = train_test_split(x, test_size = 0.25, random_state = 0, stratify = x['Outcome'])
train_X1 = train1[train1.columns[:4]]
test_X1 = test1[test1.columns[:4]]
train_Y1 = train1['Outcome']
test_Y1 = test1['Outcome']

```

```python
new_set = []
classifiers = ['Linear Svm','Radial Svm','Logistic Regression','KNN','Decision Tree']
models = [svm.SVC(kernel='linear'),svm.SVC(kernel='rbf'),LogisticRegression(),KNeighborsClassifier(n_neighbors=82),DecisionTreeClassifier()]
executable_models = []
for i in models:
    model = i
    model.fit(train_X1,train_Y1)
    executable_models.append(model)
    prediction = model.predict(test_X1)
    new_set.append(metrics.accuracy_score(prediction,test_Y1))

new_models_dataframe=pd.DataFrame(new_set, index = classifiers)   
new_models_dataframe.columns=['New Accuracy']    

```

```python
# Ще следим само по тестовата зависимост
new_models_dataframe = new_models_dataframe.merge(models_dataframe, left_index = True, right_index = True, how = 'left')
new_models_dataframe['Increase'] = new_models_dataframe['New Accuracy'] - models_dataframe['Test Accuracy']
new_models_dataframe
```


<center> <h1> Cross Validation </h1> </center>
Когато имаме небалансирани данни, е добре да тренираме и тестваме алгоритмите си на всеки клас от нашите данни. След това визмаме средното от всички резултати и представяме него.

За целта ще бъдат използвани:
* K-Fold Cross Validation, методика, която първо разделя на k на брой подчасти (k numbered subset), т.е взимаме например 1 част за тестове и тренираме алгоритъма над 4 части (k = 5). Продължаваме процеса като сменяме тестовата част за всяка итерация и тренираме алгоритъма над останалите частти. Накрая се прави средна грешка и точност над модела.

Защо да го правим? Поради това, че един алгоритъм може да underfit-ва/overfit-ва над дадени примернии тренировъчни данни. По този начин правим крос валидация и можем да достигнем до по - добро генерализиране на модела


```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

```

```python
kfold = KFold(n_splits = 10, random_state = 22)
```

```python
result_mean = []
accuracy = []
classifiers = ['Linear Svm','Radial Svm','Logistic Regression','KNN','Decision Tree']
models = [svm.SVC(kernel = 'linear'), svm.SVC(kernel='rbf'), LogisticRegression(), KNeighborsClassifier(n_neighbors = 3), DecisionTreeClassifier()]

for model in models:
    cross_result = cross_val_score(model, x[x.columns[:4]], x['Outcome'], cv = kfold, scoring = "accuracy")
    cross_result = cross_result
    result_mean.append(cross_result.mean())
    accuracy.append(cross_result)
    
new_dataframe = pd.DataFrame(result_mean, index = classifiers)
new_dataframe.columns = ['CV mean']
new_dataframe
    
```

<center> <h1> Сглобяване (Ensembling) </h1> </center>

Сглобяването на модели представлява техника, която създава множество модели и след това ги комбинира за да се получи по - добър резултат. Това е точно предимството им пред единичните модели. В случая тези единияни модели ще ги наричам базови. Сглобяването ще бъде осъществено чрез Voting Ensemble. Тази методика е достатъчно проста, за да свърши работа за конректния случай.
В основата на тази методология седи създаването на два и повече модели над подадените данни и след това се избира (извършва се voting) начин по който да се представят усреднение на предсказанията и по този начин да могат да бъдат обслужвани нови данни.

За постигане на оптимални резултати ще бъдат използвани класификатори с тежест по feature. 
От горните изследвания трите класификатора които биха били подходящи са Linaer SVM, Radial(rbf) SVM and Logistic Regression classifiers


```python
linear_svc = svm.SVC(kernel = 'linear',C = 0.1,gamma = 10,probability = True)
radial_svm = svm.SVC(kernel = 'rbf',C = 0.1,gamma = 10, probability = True)

lr = LogisticRegression(C = 0.1)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
# Linear and Radial SVM
ensemble_lin_rbf = VotingClassifier(estimators=[('Linear_svm', linear_svc), ('Radial_svm', radial_svm)], 
                       voting = 'soft', weights = [2,1]).fit(train_X1,train_Y1)
print('The accuracy for Linear and Radial SVM is:', ensemble_lin_rbf.score(test_X1, test_Y1))
```

```python
# Linear SVM with Logistic Regression
ensemble_lin_lr = VotingClassifier(estimators = [('Linear_svm', linear_svc), ('Logistic Regression', lr)], 
                       voting = 'soft', weights = [2,1]).fit(train_X1,train_Y1)
print('The accuracy for Linear SVM and Logistic Regression is:', ensemble_lin_lr.score(test_X1, test_Y1))
```

```python
# Logistic Regression with Radial SVM
ensemble_rad_lr = VotingClassifier(estimators = [('Radial_svm', radial_svm), ('Logistic Regression', lr)], 
                       voting = 'soft', weights = [1,2]).fit(train_X1, train_Y1)
print('The accuracy for Radial SVM and Logistic Regression is:', ensemble_rad_lr.score(test_X1,test_Y1))
```

```python
# Трите класификатора заедно
ensemble_rad_lr_lin = VotingClassifier(estimators = [('Radial_svm', radial_svm), ('Logistic Regression', lr),('Linear_svm',linear_svc)], 
                       voting = 'soft', weights = [2,1,3]).fit(train_X1,train_Y1)
print('The ensembled model with all the 3 classifiers is:',ensemble_rad_lr_lin.score(test_X1,test_Y1))
```
